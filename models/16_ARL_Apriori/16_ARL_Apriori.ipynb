```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
dat = pd.read_csv('../../data/Market_Basket_Optimisation.csv', header=None)
dat.head()
```

```python
dat.shape
```

```python
# apyori requires lists of transactions
# lists within a list
transactions = []
for i in range(0, 7501):
    transactions.append([str(dat.values[i, j]) for j in range(0,20)])
```

```python
transactions[0]
```

```python
# 3*7/7501 = 0.003

ap_rules = apriori(transactions=transactions,
                  min_support=3*7/7501,
                  min_confidence=0.2,
                  min_lift=3,
                  min_length=2,
                  max_length=2)
```

```python
res = list(ap_rules)
res
```

```python
# format results
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    
    return list(zip(lhs, rhs, supports, confidence, lifts))
```

```python
ap_associations = pd.DataFrame(inspect(results=res), columns=['LHS', 'RHS', 'Support', 'Confidence', 'Lift'])
ap_associations
```
